# Genetic Algorithm

```python
def crossover(parents):
    dataframe1=parents[0]
    dataframe2=parents[1]
    df1_slice = dataframe1.iloc[:18]
    df2_slice = dataframe2.iloc[18:36]
    
    df1_slice1 = dataframe1.iloc[18:36]
    df2_slice1 = dataframe2.iloc[:18]

    child1=pd.concat([df1_slice, df2_slice])
    child2=pd.concat([df1_slice1, df2_slice1])

    child1 = child1.reset_index(drop=True)
    child2 = child2.reset_index(drop=True)
    
    return child1,child2
    
```

```python
def mutation(child, probability):
    for i in range(36):
        if random.random() < probability:
            cellarray=[]
            for j in range(6):
                num=random.randint(0,381)
                cellword=classarray[0][num][0]+"/"+classarray[0][num][1]+"/"+classarray[0][num][2]+"/"+classarray[0][num][3]
                cellword=str(cellword)
                cellarray.append(cellword)
            child.iloc[i] = [cellarray[0],cellarray[1],cellarray[2],cellarray[3],cellarray[4],cellarray[5]]
    return child
```

```python
def selection(classdfarr):
    parents= random.sample(classdfarr,2)
    return parents
    
```

```python
def fitness(classdfarr):
    classsum=[]
    labsum=[]
    asum=0
    for i in range(len(classdfarr)):
        df8=classdfarr[i]
        asum=0
        for j in range(len(df8.columns)):
            a=df8.columns[j]
            asum=asum+len(df8[a])-len(df8[a].drop_duplicates())
        classsum.append(asum)
    
    return classsum
        
        
```

```python
def pop_creation(length):
    classdfarr=[]
    classcolumns=[]
    cellarray=[]
    for k in range(length):
        classcolumns=[]
        for i in range(6):
            cellarray=[]
            for j in range(len(rooms)):
                num=random.randint(0,381)
                cellword=classarray[0][num][0]+"/"+classarray[0][num][1]+"/"+classarray[0][num][2]+"/"+classarray[0][num][3]
                cellword=str(cellword)
                cellarray.append(cellword)
            classcolumns.append(cellarray)
        column_names = ['08:30-09:50','10:00-11:20','11:30-12:50','01:00-02:20','02:30-03:50','04:00-05:15']
        data_dict = dict(zip(column_names, classcolumns))
        classdf=pd.DataFrame(data_dict) 
        classdfarr.append(classdf)
    result=False
    while(result!=True):
        new_population=[]
        for num1 in range(length//2):
            parents=selection(classdfarr)
            # Crossover
            child1, child2 = crossover(parents)
            #print("parent 0",parents[0])
            #print("parent 1",parents[1])
            child1 = mutation(child1, 0.1)
            child2 = mutation(child2, 0.1)
            new_population.append(child1)
            new_population.append(child2)
            
        result1=fitness(classdfarr)
        min_num=min(result1)
        min_index=result1.index(min_num)
        new_population[0]=classdfarr[min_index]
        classdfarr=new_population
        
        if(min_num==0):
            result=True
            min_index=result1.index(min_num)
            dataframe5=new_population[0]
        
    return dataframe5  
        

```
